```python
dataset.drop(['Zomato URL','Address'],inplace=True,axis=1)
```

```python
dataset['Top Dishes'].value_counts()
```

```python
dataset['Top Dishes']=dataset['Top Dishes'].replace('Invalid','Not Available')
```

```python
dataset.reset_index(drop=True,inplace=True)
```

```python
dataset['Dining Rating'].value_counts()
```

```python
dataset[['Dining Rating', 'Dining Rating Count',
       'Delivery Rating', 'Delivery Rating Count']]=dataset[['Dining Rating', 'Dining Rating Count',
       'Delivery Rating', 'Delivery Rating Count']].replace('None','0')
```

```python
dataset
```

```python
dataset['Name of Restaurant'].value_counts()
```

```python
ck=(dataset['Name of Restaurant']=="CK's Sandwiches").sum()
ck
```

```python
ck_sandwich=dataset[dataset['Name of Restaurant']=="CK's Sandwiches"]
ck_sandwich
```

```python
dataset=dataset[dataset['Name of Restaurant']!="CK's Sandwiches"]
```

```python
dataset
```

```python
dataset['Dining Rating Count'].value_counts()
```

```python
dataset.replace(to_replace = ['None','Invalid','Does not offer Delivery','Does not offer Dining','Not enough Delivery Reviews','Not enough Dining Reviews'], value =np.nan,inplace=True)
dataset.isnull().sum()
```

```python
dataset["Dining Rating Count"].fillna(0, inplace=True)
dataset["Delivery Rating Count"].fillna(0, inplace=True)
```

```python
dataset.isnull().sum()
```

```python
zomato_df = dataset.copy() 
```

```python
zomato_df.head()
```

```python
dining_rating= zomato_df['Dining Rating'].value_counts().reset_index() 
dining_rating.rename(columns={'index':'Dining Rating','Dining Rating':'Name of Restaurant'}, inplace=True)
dining_rating
```

```python
zomato_df['Dining Rating']=pd.to_numeric(zomato_df['Dining Rating'],errors='coerce')
top_dine = zomato_df[(zomato_df['Dining Rating'] > 4.5)]
```

```python
top_dine
```

```python
#Distribution of dining ratings by histogram
plt.figure(figsize=(8,5))
sns.histplot(zomato_df['Dining Rating'],bins=20,kde=True,color='blue')
plt.xlabel('Dining Rating')
plt.ylabel('Frequency')
plt.title('Distribution of Dining Ratings')
plt.show()
```

```python
#Plotting a bar chart to show top location by restaurant counts
loc_count=zomato_df['Location'].value_counts().head(10)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=loc_count.values,y=loc_count.index,palette='viridis')
plt.xlabel('Number of Restaurants')
plt.ylabel('Location')
plt.title('Top 10 Location by Restaurant')
plt.show()
```

```python
#Plotting a scatter plot to know price for 2 vs dining rating
plt.figure(figsize=(10,6))
sns.scatterplot(x=zomato_df['Price for 2'],y=zomato_df['Dining Rating'],alpha=0.5,color='blue')
plt.xlabel('Price for 2')
plt.ylabel('Dining Rating')
plt.title('Scatter plot of price for 2 vs dining rating')
plt.show()
```

```python
#Creating a correlation heatmap for dining rating vs delivery rating
zomato_df['Delivery Rating']=pd.to_numeric(zomato_df['Delivery Rating'],errors='coerce')
```

```python
cor_matrix=zomato_df[['Dining Rating','Delivery Rating']].corr()
cor_matrix
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(cor_matrix,annot=True,cmap='coolwarm',fmt=".2f",linewidths=1)
plt.title('Correlation heatmap dining rating vs delivery rating')
plt.show()
```

```python
#Showing bar chart for top restaurants with high dining rating
top_restaurant=zomato_df.nlargest(10,'Dining Rating')[['Name of Restaurant','Dining Rating']]
top_restaurant
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='Dining Rating',y='Name of Restaurant',data=top_restaurant,palette='viridis')
plt.xlabel('Dining Rating')
plt.ylabel('Name of Restaurant')
plt.title('Top 10 Restaurant with high dining rating')
plt.show()
```
